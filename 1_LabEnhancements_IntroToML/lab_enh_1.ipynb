In this lab, we will be training a basic machine learning model for predicting numerical outputs using the scikit-learn's libary.

<font size = 5 color='plainyellow'><u><i>Scikit-learn</i></u></font>

```python
# Note: In Juypter, libraries can be imported later for code clarity.

import pandas as pd

# Step 1: Load in the dataset
file_path = "Housing_prices_train_syn.csv"
home_data = pd.read_csv(file_path)
```

```python
# Step 2: Inspect or study the chosen dataset
X = home_data
X.head()
```

As we inspect and study the dataset, pay attention to the variables (column names and their values).

Why?
- They will help us in determining whether our questions can be answered.
- Help us explore questions and relationships between variables.

These variables are known as features, in this case, features of many different houses in a dataset. Likewise, the data for these features are also different for each varying houses.

To demostrate how it works, we can aim for a simple question.

Question: <span style="color:deepskyblue; font-weight:bold;">What are the actual home values from the dataset?</span>



```python
# Step 3: Pick the data features you think its suitable for your question.

feature_names = ["LotArea","YearBuilt","1stFlrSF","2ndFlrSF","FullBath","BedroomAbvGr","TotRmsAbvGrd"]
X = home_data[feature_names]

# Step 4: set your prediction target that is already accurate.
prediction_target = ["SalePrice"]
y = home_data[prediction_target]
```

```python
# Step 5: Review your chosen data features from the dataset

print(X)
X.head()

print(y)
y.head()
```

```python
# Step 6: Fit the chosen dataset into the ML Algorithm
from sklearn.tree import DecisionTreeRegressor

# Decision Tree Algorithm is used, randomness in choosing data for prediction consistency
model_1 = DecisionTreeRegressor(random_state=1)

# X = Features chosen to predict
# y = target prediction
model_1.fit(X, y)
```

```python
# Step 7: Make a prediction with your dataset in the ML Algorithm

predictions = model_1.predict(X.head())
predictions
```
